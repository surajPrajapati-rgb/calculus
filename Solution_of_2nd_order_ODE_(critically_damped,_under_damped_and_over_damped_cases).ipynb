#Second Order ODE Homogenous Equation Solutions:

A second-order ODE has the form:

$$Ax'' + Bx' + Cx = 0$$
The solution of the given differential equition is given by
$$\lambda=\frac{-B\pm\sqrt{B^{2}-4AC}}{2A}$$
 
#Kinds of solution:

1. Under-Damped Case:

 When $B^{2}-4AC<0,\,\lambda_{1}\,\,and\lambda_{2}$ are complex conjugate in pairs., 

2. Critically Damped Case:

 When  $B^2 - 4AC = 0$, $\lambda_{1}\,\,and\lambda_{2}$ are real and equal


3. Over-Damped Case:

 When $B^{2}-4AC>0,\,\lambda_{1}\,\,and\lambda_{2}$ are real and distinct


Finding the roots of Given Differential Equation.

Initial conditions:-

$x(0)	=x_{0}, 
x^{'}(0)	=v_{0}$

$$x=D_{1}x_{1}+D_{2}x_{2}$$
$$x(t)=D_{1}x_{1}(t)+D_{2}x_{2}(t)$$
$$x(0)={{D_{1}e^{\lambda_{1}(0)}+D_{2}e^{\lambda_{2}(0)}}}$$
$$x_{0}=x(0)=D_{1}+D_{2}$$
$$x^{'}(t)=D_{1}x_{1}^{'}(t)+D_{2}x_{2}^{'}(t)$$
$$x^{'}(t)=D_{1}\lambda_{1}e^{\lambda_{1}t}+D_{2}\lambda_{2}e^{\lambda_{2}t}$$
$$v_{0}(0)=x^{'}(0)=D_{1}\lambda_{1}+D_{2}\lambda_{2}$$


$$\lambda=\frac{-B\pm\sqrt{B^{2}-4AC}}{2A}{{=-\frac{B}{2A}\pm\sqrt{\frac{B^{2}}{4A^{2}}-\frac{C}{A}}=-\alpha\pm\beta}}{ {\beta=\sqrt{\frac{B^{2}}{4A^{2}}-\frac{C}{A}}}}where\,D1=\frac{v_{o}-x_{o}\lambda_{2}}{\lambda_{1}-\lambda_{2}}\,\,\&\,\,D2=\frac{v_{o}-\lambda_{1}x_{o}}{\lambda_{2}-\lambda_{1}}$$


Case(i):-For Over Damping

$$x=D_{1}e^{(-\alpha+\beta)t}+D_{2}e^{(-\alpha-\beta)t}$$

Case(ii):-For Critical Damping

$$x=(x_{0}+(v_{0}-\lambda x_{0})t)e^{\frac{-B}{2A}t}$$

Case(iii):-For Under Damping

$$x=D_{1}e^{(-\alpha+i\beta)t}+D_{2}e^{(-\alpha-i\beta)t}$$

$$\Rightarrow x=e^{-\alpha t}(D_{1}e^{(i\beta)t}+D_{2}e^{(-i\beta)t})$$

```python
import math 
import matplotlib.pyplot as plt
```

```python

A = float(input('A :'))
B = float(input('B :'))
C = float(input('C :'))
x0 = float(input('x0 :'))
v0 = float(input('v0 :')) 
n = int(input("input no of observation : "))
```

```python
alpha = B/(2*A)
beta = math.sqrt(abs((B**2/(4*A**2) - C/A)))
# Discriminant value of second order D.E.
D = B**2-4*A*C

# Checking for types of Damping
def damping_type(A,B,C,x0,v0):
  if D > 0:
    return 'Overdamping'
  elif D == 0:
    return 'critically damping'
  elif D < 0:
    return 'underdamping'
type = damping_type(A,B,C,x0,v0)
```

```python
tlist = []            # accu for x-axis
xlist = []          # accu for y-axis

for i in range(n) :
  # Appending value for x-axis
  tlist.append(i/10) 

for t in tlist:
# In the case of Overdamping
  if D > 0 : 
    # Value of lembda 
    lem1 = (-B + math.sqrt(D))/(2*A)
    lem2 = (-B - math.sqrt(D))/(2*A)

    # Cofficiant of x1 and x2 in general solution 
    D1 = (v0 - lem2*x0)/(lem1-lem2)
    D2 = (v0 - lem1*x0)/(lem2-lem1)

    # Two Absolute solution
    x1 = D1*math.exp(lem1*t)
    x2 = D2*math.exp(lem2*t)

    # General solution
    x = x1*D1 + x2*D2

# In the case of Critically Damping
  elif D == 0 :    
    lembda = -B/(2*A)
    x = (x0 + (v0 - lembda*x0)*t)*math.exp((-B*t)/(2*A))

# In the case of Underdamping
  elif D < 0:       
    x = math.exp(-alpha*t)*(x0*math.cos(beta*t) + math.sin(beta*t)*(v0+alpha*x0)/(beta))

  # Appending value(x) for y-axis corresponding to value in tlist  
  xlist.append(x)         
```

```python
# import matplotlib.pyplot as plt
plt.plot(tlist,xlist)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title(type)
plt.show()
```
